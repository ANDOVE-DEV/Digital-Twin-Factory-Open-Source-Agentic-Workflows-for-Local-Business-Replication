# ðŸŒ± GreenAI: Modellazione della Crescita Botanica

Questo modello scientifico simula lo sviluppo di una pianta basandosi su variabili ambientali reali.

```python
import numpy as np
import matplotlib.pyplot as plt

def logistic_growth(t, K=100, r=0.01, P0=1):
    # Modello di crescita logistica
    return K / (1 + ((K - P0) / P0) * np.exp(-r * t))

days = np.linspace(0, 500, 100)
growth = [logistic_growth(d) for d in days]

plt.figure(figsize=(10, 6), facecolor='#040807')
ax = plt.axes()
ax.set_facecolor('#0a1411')
plt.plot(days, growth, color='#10b981', linewidth=3)
plt.title("Curva di MaturitÃ  Biologica", color='white')
plt.xlabel("Tempo (Ore di Simulazione)", color='white')
plt.ylabel("MaturitÃ  (%)", color='white')
plt.tick_params(colors='white')
plt.grid(alpha=0.1)
plt.show()
```

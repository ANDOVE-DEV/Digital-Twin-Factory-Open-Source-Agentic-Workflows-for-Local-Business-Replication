# üè≠ OpenFactoryTwin: Modello di Simulazione Energetica

Questo notebook modella la relazione tra velocit√† di produzione e consumo energetico per ottimizzare la sostenibilit√† della linea Mini Factory.

```python
import numpy as np
import matplotlib.pyplot as plt

def model_consumption(speed):
    # Consumo base + crescita quadratica con la velocit√†
    base_load = 0.5
    dynamic_load = 0.2 * (speed ** 2)
    return base_load + dynamic_load

speeds = np.linspace(0, 10, 100)
consumptions = [model_consumption(s) for s in speeds]

plt.figure(figsize=(10, 6))
plt.plot(speeds, consumptions, label="Consumo (kW)")
plt.xlabel("Velocit√† (Unit√†/ora)")
plt.ylabel("Potenza Assorbita (kW)")
plt.title("Relazione Velocit√†-Consumo per Macchina M1")
plt.grid(True)
plt.legend()
plt.show()
```
